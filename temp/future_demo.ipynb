```python
import pandas as pd
from JuRunPlatform import StrategyPlatform
import datetime
import matplotlib.pyplot as plt
```

```python
pd.read_csv("future_continuous_info.csv")                           #encoding="gb2312"
```

```python
sp = StrategyPlatform()
# 设置要回测的品种（必须）
#codes="RB0.SHF,JM0.DCE,J0.DCE,HC0.SHF,I0.DCE".split(',')#螺纹，焦煤，焦炭，热卷，铁矿
sp.SetFutureList("RB0.SHF".split(','))  # 期货连续 RB螺纹钢
```

可调用的因子列表：
'open_N', 'close_N', 'high_N', 'low_N', 'amt', 'pct_chg', 'volume'

```python
# 要操作的日期（必须）#开始日期20100104
sp.SetDate(20100104, 20191008)
# 策略begin
close = sp.LoadData('future','close_N')
#openn =sp.LoadData('open_N')
#获取交易日序列
dates= sp.btp.date_num_list_trading
dates
```

```python
datestr=[]
#日期序列转化日期格式
for da in dates:
    time_format = datetime.datetime.strptime(str(int(da)), '%Y%m%d')
    datestr.append(time_format)
```

```python
df1=pd.DataFrame(close,columns=["close"],index=datestr)
df1["ret"]=df1["close"]-df1["close"].shift(1)
df1["mean5"]=df1["close"].rolling(5).mean()
df1["mean20"]=df1["close"].rolling(20).mean()
df1
```

```python
alpha = sp.LoadZeroAlpha()
alpha
```

```python
for i in range(len(alpha)-1):
    if(df1.mean5.iloc[i]>df1.mean20.iloc[i]):  #短均线上穿越长均线，开多仓
        alpha[i+1]=1#注意i+1
    elif df1.mean5.iloc[i]<=df1.mean20.iloc[i]:#短均线下穿长均线，开空仓
        alpha[i+1]=-1
```

```python
df1["alpha"]=alpha
df1["pnl"]=df1.ret*df1.alpha
df1["cumpnl"]=df1.pnl.cumsum()
df1["benchmark"]=df1.ret.cumsum()
#df1.cumpnl.plot()
df1
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df1.cumpnl,label='strategy')
plt.plot(df1.benchmark,label="benchmark")
plt.legend(['strategy','benchmark'])
```
