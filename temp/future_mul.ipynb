```python
sp = StrategyPlatform()
# 设置要回测的品种（必须）
codes="RB0.SHF,JM0.DCE,J0.DCE,HC0.SHF,I0.DCE".split(',')#螺纹，焦煤，焦炭，热卷，铁矿
sp.SetFutureList(codes)  # 期货连续 
```

```python
sp.SetDate(20180104, 20190915)
# 策略begin
close = sp.LoadData('future','close_N')
#openn =sp.LoadData('future',open_N')

#获取交易日序列
dates= sp.btp.date_num_list_trading
datestr=[]
#日期序列转化日期格式
for da in dates:
    time_format = datetime.datetime.strptime(str(int(da)), '%Y%m%d')
    datestr.append(time_format)
```

```python
def transalpha(tem):   
    alpha=tem.copy()
    abs1=tem.abs()
    absmax=abs1.max()
    target=tem[abs1==absmax]
    target_index=target.index
    for i in alpha.index:
        if  i==target_index:
            if target.values >0:
                alpha.loc[i]=-1
            elif target.values<0:
                alpha.loc[i]=1
        else:
            alpha.loc[i]=0
    return(alpha)
```

```python
transalpha(dfend.iloc[5,:])
```

```python
alpha = sp.LoadZeroAlpha()
```

```python
for i in range(len(alpha)-1):
    alphatem=transalpha(dfend.iloc[i,:])
    alpha[i+1,:]=alphatem.values   
```

```python
def performance(df_close,alpha,cash_init=10000000):
    ret=df_close-df_close.shift(1)
    base=cash_init/df_close
    dfen=alpha*base
    pnl=dfen.sum(axis=1)
    plt.figure(figsize=(15, 5))
    plt.plot(pnl.cumsum(),label='strategy')
```

```python
performance(df_close,alpha,cash_init=10000000)
```
